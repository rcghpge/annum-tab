```python
import jax
import jax.numpy as jnp

# Parameters
key = jax.random.PRNGKey(0)
X = jax.random.normal(key, (100, 10))
y = jax.random.normal(key, (100, 1))

# Initialize weights
W1 = jax.random.normal(key, (10, 32)) * 0.1
b1 = jnp.zeros((1, 32))
W2 = jax.random.normal(key, (32, 1)) * 0.1
b2 = jnp.zeros((1, 1))

# Activation
def relu(x):
    return jnp.maximum(0, x)

# Forward pass function
def forward(params, X):
    W1, b1, W2, b2 = params
    A1 = relu(jnp.dot(X, W1) + b1)
    out = jnp.dot(A1, W2) + b2
    return out

params = [W1, b1, W2, b2]
output = forward(params, X)
print("‚úÖ JAX MLP forward pass output shape:", output.shape)
```
