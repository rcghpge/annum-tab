# test uv stack on bsd

```python
import shutil

if shutil.which("uv"):
    print("✅ uv is installed and available on PATH.")
else:
    print("❌ uv not found in PATH.")
```

```python
import subprocess

def list_with_pip():
    result = subprocess.run(
        ["uv", "pip", "list"],
        check=True,
        text=True,
        capture_output=True
    )
    print(result.stdout)

# Print results
list_with_pip()
```
